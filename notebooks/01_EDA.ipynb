# 01 — Exploratory Data Analysis

Inspect raw experiment inputs and sanity-check the data before running any statistical tests.

---
**What this notebook covers:**
- Loading experiment data from CSV
- Inspecting conversion rates for both variants
- Checking sample sizes and identifying imbalance
- Flagging obvious data quality issues

```python
# ── path setup — run this cell first ─────────────────────────────────────────
import sys
from pathlib import Path

ROOT = Path().resolve().parent   # works when notebook is in notebooks/
SRC  = ROOT / 'src'
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

DATA = ROOT / 'data' / 'sample_experiment.csv'
print('Project root:', ROOT)
print('Data file:   ', DATA)
```

```python
import pandas as pd
from ab_testing_framework.data_loader import load_data

# Load the sample dataset
df = pd.read_csv(DATA)
print('Raw CSV:')
df
```

```python
# Compute and display conversion rates
df['rate_a'] = df['conversions_a'] / df['visitors_a']
df['rate_b'] = df['conversions_b'] / df['visitors_b']
df['absolute_lift'] = df['rate_b'] - df['rate_a']
df['relative_lift'] = df['absolute_lift'] / df['rate_a']

print(df[['visitors_a','conversions_a','rate_a',
          'visitors_b','conversions_b','rate_b',
          'absolute_lift','relative_lift']].to_string())
```

```python
# Parse through the framework validator
experiment = load_data(DATA)
print('Validated ExperimentInput:')
print(f'  Control:   {experiment.visitors_a:,} visitors, {experiment.conversions_a:,} conversions '
      f'({experiment.conversions_a / experiment.visitors_a:.2%})')
print(f'  Treatment: {experiment.visitors_b:,} visitors, {experiment.conversions_b:,} conversions '
      f'({experiment.conversions_b / experiment.visitors_b:.2%})')
print(f'  Alpha:     {experiment.alpha}')
```

### Observations

- Both groups have equal sample sizes (10,000) — no imbalance issues.
- Control rate: **4.50%**, Treatment rate: **5.20%**.
- Absolute lift: **+0.70 pp**, relative lift: **+15.6%**.
- No missing values or obvious data quality problems.
